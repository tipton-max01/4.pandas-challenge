```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
#Display the total number of players
player_count = len(purchase_data["SN"].unique())
total_players = pd.DataFrame({"Total Players": [player_count]})
total_players
```

## Purchasing Analysis (Total)

```python
#Run basic calculations to obtain number of unique items, average price, etc.

#unique items
unique_items = len(purchase_data["Item Name"].unique())

#average price
average_price = purchase_data["Price"].mean()

#number of purchases
number_purchases = purchase_data["Price"].count()

#total income from purchases
total_revenue = purchase_data["Price"].sum()

#Optional: give the displayed data cleaner formatting
```

```python
#Create a summary data frame to hold the results
#Display the summary data frame
purchasing_analysis = pd.DataFrame({"Unique Items": [unique_items],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [number_purchases],
                                    "Total Revenue": [total_revenue],
                                   })
purchasing_analysis.style.format({"Average Price": "${:,.2f}", 
                                  "Total Revenue": "${:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
max_age = purchase_data["Age"].max()
min_age = purchase_data["Age"].min()
print(f"The maximum age of our group is: {max_age}")
print(f"The minimum age of our group is: {min_age}")
```

```python
# Create the bins in which Data will be held
# Bins are 0, 59.9, 69.9, 79.9, 89.9, 100.   
#bins = [0, 59.9, 69.9, 79.9, 89.9, 100]

age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Categorize the existing players using the age bins
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_data

#Create Age Demographics dataframe by removing duplicates from purchase_data
age_demographics = purchase_data.drop_duplicates(subset = ["SN"], inplace = False, keep = "first")
age_demographics

age_count = age_demographics.groupby(["Age Ranges"]).count()["SN"]
age_count

total_people = age_count.sum()
total_people

age_count_percentages = age_count / total_people
age_count_percentages

age_demo_df = pd.DataFrame({"Age Bins": group_names,
                            "Count per Bin": age_count,
                            "Percentage of Count": age_count_percentages,                    
                           })
age_demo_df["Percentage of Count"] = age_demo_df["Percentage of Count"].map("{:,.2%}".format)
age_demo_df
```
